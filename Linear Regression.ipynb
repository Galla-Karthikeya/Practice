```python
df[['CHAS', 'RAD']] = df[['CHAS', 'RAD']].astype('int64')
```

```python
def outlier(df, i):
    q3 = df[i].quantile(.75)
    q1 = df[i].quantile(.25)
    IQR = q3 - q1
    uf = q3 + 1.5*IQR
    lf = q1 - 1.5*IQR
    def map_value(value):
        if(value >= uf):
            return uf
        elif(value <= lf):
            return lf
        else:
            return value
    df[i] = df[i].apply(map_value)
    
def outlier_deletion(df):
    for col in df.columns:
        print(col)
        outlier(df, col)

outlier_deletion(df)    
```

```python
for i in df.columns:
    if(i not in ['CHAS']):
        sns.boxplot(y = df[i])
        plt.show()
```

```python
# Define a custom color palette
custom_palette = sns.color_palette("husl", len(df['RAD'].unique()))


for i in df.columns:
    if(i not in  ['Target', 'CHAS']):
        sns.scatterplot(x = df['Target'], y = df[i], hue = 'RAD', data = df, palette = custom_palette)
        plt.xlabel('Target')
        plt.ylabel(i)
        plt.show()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_sc = sc.fit_transform(x)
x_sc = pd.DataFrame(x_sc, columns = x.columns)
x_sc.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, test_size = 0.2, random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_predict = lr.predict(x_test)
y_predict
```

```python
# we will observe it in scatter plot
```

```python
sns.scatterplot(x = y_test, y = y_predict)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.show()
```

```python
print("The Slope is ", lr.coef_)
print("The Intercept line is : ", lr.intercept_)
```

```python
min_val = (y_predict.min()+ y_test.min())//2
max_val = (y_predict.max() + y_test.max())//2
print(min_val, max_val)
```

```python
sns.scatterplot(x = y_test, y = y_predict, color = 'green')
plt.plot([min_val, max_val], [min_val, max_val], color = 'maroon', linestyle = '--')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
MSE = mean_squared_error(y_predict, y_test)
MSE
```

```python
MAE = mean_absolute_error(y_predict, y_test)
MAE
```

```python
r2score = r2_score(y_predict, y_test)
print('Accuracy/ R2_Score: ', r2score)
```

```python
RMSE = np.sqrt(MSE)
RMSE
```
