```python
def calculator(num1, num2, symbol):
    if(symbol == '+'):
        return num1+num2
    elif(symbol == '-'):
        return num1 - num2
    elif(symbol == '*'):
        return num1 * num2
    elif(symbol == '/'):
        return num1 /num2
    else:
        return "Selected Wrongly select one of these \'add, sub, mul, div\'"
        
```

```python
calculator(534, 656, "+")
```

```python
inputs = [
    gr.components.Number(label = 'Number1'),
    gr.components.Number(label = 'Number2'),
    gr.components.Dropdown(choices = ['+', '-', '*', '/'], label = 'operations')
]

outputs = gr.components.Number(label = 'Result')

gr.Interface(fn = calculator, inputs = inputs, outputs= outputs).launch()
```

```python
def start(name, morning_is, temp):
    msg = 'Good Morning' if morning_is == 'Morning' else 'Good Evening'
    greeting = '%s %s. \nIt is %s degrees today' %(msg, name, temp)
    cels = f'The Temperature is {round((temp - 32)*5/9, 2)} (C)'
    return greeting, cels

inputs = [
    gr.components.Text(label = 'Name'),
    gr.components.Radio(choices = ['Morning', 'Evening'], label = 'parts of the day'),
    gr.components.Slider(-130, +150, label="Temperature (F)")
]

outputs = [
    gr.components.Text(label = 'Details'),
    gr.components.Text(label="Celsius")
]
```

```python
face = gr.Interface(fn = start, inputs = inputs, outputs = outputs)
```

```python
face.launch(share = True)
```

```python
import gradio as gr

def Details(name, gender, age, status, clg, state):
    gender = 'Male' if gender == 'Male' else 'Female'
    status = 'Single' if status else 'Commited'
    return 'Name: %s \nGender: %s \nAge: %s \nStatus: %s \nCollage: %s \nState: %s'%(name, gender, age, status, clg, state)

result = gr.Interface(fn = Details, 
             inputs = [gr.components.Text(placeholder = 'Name of a Student', label = "Student Name"),
                       gr.components.Radio(choices = ['Male', 'Female'], label = "Gender"),
                       gr.components.Slider(0, 100, label = 'AGE'),
                       gr.components.Checkbox(label = 'Single'),
                       gr.components.Textbox(lines = 2, label = "Collage"),
                       gr.components.Textbox(label = 'State')
                      ],
             outputs = gr.components.Text(label = 'Details'))
```

```python
result.launch(share = True)
```

```python
import random
import gradio as gr

def game1(input):
    message = random.choice(['Rock', 'Paper', 'Scissors'])
    if(input == 'Rock'):
        if(message == 'Scissors'):
            game = "You Won"
        elif(message == 'Rock'):
            game = 'It\'s Draw'
        else:
            game = "You Lost"
    elif(input == 'Paper'):
        if(message == 'Scissors'):
            game = 'You Lost'
        elif(message == 'Rock'):
            game = 'You Won'
        else:
            game = "It's Draw"
    else:
        if(message == 'Scissors'):
            game = 'It\'s Draw'
        elif(message == 'Rock'):
            game = 'You Lost'
        else:
            game = "You Won"
    return message, game

def game2(Your_choice, Your_number):
    c_number = random.choice([1, 2, 3, 4, 5])
    total = Your_number + c_number
    if(total %2 == 0):
        if(Your_choice == 'Even'):
            c_choice = 'Odd'
            game = 'You Won'  
        else:
            c_choice = 'Even'
            game = 'Computer Won'
            
    elif(total %2 != 0):
        if(Your_choice == 'Odd'):
            c_choice = 'Even'
            game = 'You Won'
        else:
            c_choice = 'Odd'
            game = 'Computer Won'

    return c_number, c_choice, total, game
    

inputs1 = gr.components.Text(label = 'You Turn', placeholder = 'You Choice')
outputs1 = [
    gr.components.Text(label = 'Computer Turn', placeholder = 'Computer Choice'),
    gr.components.Text(label = 'Result', placeholder = 'Who Won')
]

inputs2 = [
    gr.components.Text(label = 'Your Turn', placeholder = 'Your Choice'),
    gr.components.Number(label = 'Your Number')
]
outputs2 = [
    gr.components.Number(label = 'Computer Number'),
    gr.components.Text(label = 'Computer Turn', placeholder = 'Computer Choice'),
    gr.components.Number(label = 'Total Sum'),
    gr.components.Text(label = 'Result', placeholder = 'Result')
]

App1 = gr.Interface(fn = game1, inputs = inputs1, outputs = outputs1, title="Rock Paper Scissors Game", description="Enter Rock, Paper, or Scissors and see if you can beat the computer!")
App2 = gr.Interface(fn = game2, inputs = inputs2, outputs = outputs2, title="Odd Even Game", description="Enter Odd or Even and Choose a Number and see if you can beat the computer!")


result = gr.TabbedInterface([App1, App2],  ['APP1', 'APP2'])
result.launch()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

subject = []
value = []
def data(sub, val):
    subject.append(sub)
    value.append(val)
    return f"Subject: {sub} \nValue: {val}"

def plot():
    x = ['Maths', 'Physics', 'Chemistry', 'Computer Science', 'AIML']
    y = [43, 13, 2, 45, 50]
    df = pd.DataFrame()
    df['Subjects'] = x
    df['Values'] = y
    plt.figure(figsize = (10, 5))
    plot = plt.bar(df['Subjects'], df['Values'], color = 'red')
    plt.xlabel('Subjects')
    plt.ylabel('Values')
    return plt.gcf()

output = gr.Plot(label="Subject Values Bar Chart")

input = [
    gr.components.Text(label = 'Subject'),
    gr.components.Number(label = 'Values')
]

outputs = gr.components.Text(label = 'RESULT')



data = gr.Interface(fn = data, inputs = input, outputs = outputs)
graph = gr.Interface(fn = plot, inputs = None, outputs = output)

result = gr.TabbedInterface([data, graph], ['Data', 'Graph']).launch(share = True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
data_df = load_iris()
data_df
```

```python
df = pd.DataFrame(data_df.data, columns = data_df.feature_names)
df.head()
```

```python
df['Target_names'] = df['Target'].apply(lambda x: data_df.target_names[x])
```

```python
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
x = df.iloc[:, :-2]
y = df.iloc[:, -2]
```

```python
lr.fit(x, y)
```

```python
lr.score(x, y)
```

```python
new_data = np.array([[5.1, 3, 4, 0.2]])
value = lr.predict(new_data)
print(value[0])
print(data_df.target_names[value][0])
```

```python
import gradio as gr

def predict(sepal_length, sepal_width, petal_length, petal_width):
    value = lr.predict([[sepal_length, sepal_width, petal_length, petal_width]])
    if(value == 0):
        img_url = r'https://www.growildnursery.co.uk/uploads/3/1/3/9/31399963/s310177535295035342_p464_i1_w2480.jpeg'
    elif(value == 1):
        img_url = r'http://www.beautifulflowerpictures.com/blog/wp-content/uploads/2012/11/iris_virginia_southern_blue_flag_9336.jpg'
    else:
        img_url = r'https://i.pinimg.com/originals/7d/3e/1c/7d3e1c646a5900a5312851db7c763f0d.jpg'
    return data_df.target_names[value][0], img_url

inputs = [
    gr.components.Number(label = 'sepal length (cm)'), 
    gr.components.Number(label = 'sepal width (cm)'),
    gr.components.Number(label = 'petal length (cm)'),
    gr.components.Number(label = 'petal width (cm)')
]
outputs = [
    gr.components.Text(label = 'Predicted Output'),
    gr.Image(label = 'Image')
]

gr.Interface(fn = predict, inputs = inputs, outputs= outputs).launch()
```
